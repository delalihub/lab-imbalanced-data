```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
dataa =pd.read_csv('customer_churn.csv')
dataa.head(50)
```

```python
dataa_drop = dataa.drop(['tenure', 'SeniorCitizen', 'MonthlyCharges', 'Churn'], axis=1)
dataa_drop.columns
data = dataa[['tenure', 'SeniorCitizen', 'MonthlyCharges', 'Churn']]
data
```

# It is convenient to covert categorical values in target columns into booleans although they could be kept

```python
data["Churn"].unique()
```

```python
def boolean(x):
    if x=='No':
        return 0
    else:
        return 1
data['Churn'] = list(map(boolean, data['Churn']))
```

```python
data
```

```python
data["Churn"].value_counts()
```

```python
#Splitting into "data" into X and y
# X/y split
X = data.drop('Churn',axis = 1)
y = data['Churn']
```

# When target values are categorical, doont use them in checking multicoleanity

```python
# For logistic regression we drop the y value since its not numerical.The "y" is categorical or bolean so not good
#for multicoleanity. so ew drop "Y" and keep "X"
correlations_matrix = X.corr()
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

# Standardize numerical independent variables with (Minimize Scaler or Standard scaler)

```python
# 1 But Before standardizing, we will need to split
# here we start scaling, so need to do train-test-split before
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0) # default fraction is .2
```

```python
# all features are numeric, so no need to split into _num and _cat
transformer = StandardScaler().fit(X_train)
X_train_scaled = pd.DataFrame(transformer.transform(X_train),columns=X.columns)
# because this is the only tranformation we apply to the numerics, 
# we can immediately transform the X_test as well
X_test_scaled = pd.DataFrame(transformer.transform(X_test),columns=X.columns)
X_train_scaled.head()
```

```python
y_train.head(5)
```

# y indexes changed after split, training; reset them with (.reset_index(drop=True))

```python
y_train = y_train.reset_index(drop=True) 
```

```python
# and for y_test
display(y_test)

y_test = y_test.reset_index(drop=True) 
y_test
```

# Linear relation on logit

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_scaled, y_train)
```

# predicted probability of  "X_train_scaled"

```python
LR.predict_proba(X_train_scaled)
```

# Visualising some columns of the "X_train_scaled"

```python
logits = np.log(1/LR.predict_proba(X_train_scaled)[:,1]-1)
plt.scatter(logits,X_train_scaled['tenure']) #shows positive correlation
```

```python
plt.scatter(logits,X_train_scaled['SeniorCitizen']) # What? how do I describe this
```

# Checking the balance/Imbalance metrics of my target variable (Churn)

```python
count_classes = data['Churn'].value_counts()
count_classes
count_classes.plot(kind = 'bar') # Major indication of Imbalance from the target columns categoris
```

# My Model results

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_scaled, y_train) # trained and fitted set with model
LR.score(X_test_scaled, y_test) # No neeed to be retrained nor fitted
```



# How good was my model using logistic regression

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

pred = LR.predict(X_test_scaled)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```

# precision :-  Compares the True Prediction with "Predicted values"
# recall:- compares the True Prediction with the " Actual Values"
# fi: - is the mean of "Precision" and "recall"

```python
#The "recall" shows  about 54% falurure to identify ("churn") cases!
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```

### Since "recall" gap(100%-recall%) is that huge, I can Oversample, under-sample or smot. I choose the first 2

```python
# oversampling / undersampling is only to be done on the TRAINING set
# our test set still must reflect reality!
train = pd.concat([X_train_scaled, y_train],axis=1)
train.head()
```

# USING OVERSAMPLING,

```python
no_churn = train[train['Churn']==0]
yes_churn = train[train['Churn']==1]
```

```python
from sklearn.utils import resample
# To oversample, use (yes_churn) which is the smaller or in  minority

yes_churn_oversampled = resample(yes_churn, #<- sample from here
                                    replace=True, #<- we need replacement, since we don't have enough data otherwise
                                    n_samples = len(no_churn),#<- make both sets the same size
                                    random_state=0)
```

```python
# both sets are now of a reasonable size
display(no_churn.shape)
display(yes_churn_oversampled.shape)
yes_churn_oversampled.head(20)
```

# Oversampled data needs to be trained

```python
train_oversampled = pd.concat([no_churn,yes_churn_oversampled],axis=0)
train_oversampled.head()
```

```python
y_train_over = train_oversampled['Churn'].copy()
X_train_over = train_oversampled.drop('Churn',axis = 1).copy()
```

# Model Improved after oversampling referece: "recall, Precision"

```python
# Our Logistic Regression has improved substantially!
# especially at detecting instances of "Churn"

LR_over = LogisticRegression(random_state=0, solver='lbfgs')
LR_over.fit(X_train_over, y_train_over)
pred = LR_over.predict(X_test_scaled)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```

```python
confusion_matrix(y_test,pred)
```

# USING SMOT

```python
# you can find documentation here: https://imbalanced-learn.readthedocs.io/en/stable/generated/imblearn.over_sampling.SMOTE.html
# by deafult, takes a 5-neighbour KNN to build a new point, experimenting led us to choose 3 here
sm = SMOTE(random_state=100, k_neighbors=3)
X_train_SMOTE,y_train_SMOTE = sm.fit_resample(X_train_scaled,y_train)
```

```python
X_train_SMOTE.shape
```

# SMOT IMPROVED THE MODEL

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_SMOTE, y_train_SMOTE)
pred = LR.predict(X_test_scaled)

print("precision: ",precision_score(y_test,pred))
print("recall: ",recall_score(y_test,pred))
print("f1: ",f1_score(y_test,pred))
```
